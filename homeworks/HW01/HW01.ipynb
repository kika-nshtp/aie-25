```python
import numpy as np
import math


# конвертация градусов в радианы
degree = 90
radian = math.radians(degree)
print(f'Конвертация радиан в градусы и обратно\nС помощью библиотеки math:\n{degree} градусов = {radian} радиан')

# конвертация радиан в градусы
radian = math.pi
degree = math.degrees(radian)
print(f'{radian} радиан = {degree} градусов')

# способ "вручную"
degree = 45
radian = degree/180*math.pi
print(f'Вручную:\n{degree} градусов = {radian} радиан')
radian = 1
degree = radian*180/math.pi
print(f'{radian} радиан = {degree} градусов')


# работа с библиотекой numpy
# примеры создания массивов
array1 = np.array([1, 34, 5.8, 92, -10])
array2 = np.array([[-3, 65, 87.5],[43, -59, 56.4]]) 
array3 = np.zeros((2,4))
array4 = np.ones((3,5))
array5 = np.eye(4)
array6 = np.empty((3,4))
array7 = np.arange(10,31,5)
print(f'\nПримеры создания массивов\nОдномерный массив \n{array1}')
print(f'Двумерный массив \n{array2}')
print(f'Использование функции zeros\n{array3}')
print(f'Использование функции ones\n{array4}')
print(f'Использование функции eye\n{array5}')
print(f'Использование функции empty\n{array6}')
print(f'Использование функции arange\n{array7}')

# демонстрация атрибутов
print(f'\nВывод формы и типа данных для двумерного массива\n{array2.shape}\n{type(array2)}')

# примеры индексирования и срезов
print(f'\nПолучение отдельного элемента\n{array6[1,2]}')
print(f'Извлечение подмассива с помощью среза\n{array6[[1,2]]}')
# примеры векторных операция и базовых статистик
print(f'\nПоэлементное сложение и умножение массивов\n{array1+array7}\n\n{array1*array7}')
print(f'Вычисление суммы элементов массива, средней величины, максимального элемента массива\n{np.sum(array2)}, {np.mean(array6)}, {array1.max()} ')
```
